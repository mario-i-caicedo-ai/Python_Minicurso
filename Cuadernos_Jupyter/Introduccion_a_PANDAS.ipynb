# PANDAS

* Pandas es una de las tres bibliotecas fundamentales para analítica

* Pandas permita construir una estructura de datos llamada: **DATA FRAME**

* Desde un punto de vista sumamante simplificado, una DATAFRAME es una tabla.

* PANDAS utiliza la biblioteca numpy (numeric python)



```
import pandas as pd

import numpy as np
```

### La estructura de datos fundamental de PANDAS es denominada:

### **Pandas series**

* Una serie pandas posee tres elementos: 

  * Índice
  * Nombre y
  * Una lista (los datos)  

# Ejemplos de Creación de PANDAS SERIES

https://es.wikipedia.org/wiki/Eurocopa

```
# create a series from a simple array
Paises = np.array(['Alemania', 'España', 'Italia', 'Francia','Rusia'])
  
Ganadores_Top = pd.Series(Paises,name='Ganadores de Eurocopa')
print(Ganadores_Top)
```

* En este ejemplo queremos destacar la definición del índice como parte de la serie.

```
N_Campeonatos = pd.Series([3,3,2,2,1], index =[5, 6, 7, 8,9],name='Número de Copas')
N_SubCampeonatos = pd.Series([3,1,2,1,3], index =[10, 20, 30, 40,50],name='Número de Copas')
print(N_Campeonatos)
```

```
Fechas_Campeon = pd.Series(['1972, 1980, 1996','1964, 2008, 2012','1968, 2020','1984, 2000','1960'], index =[3, 3, 2, 2,1],name='Año_1')
Fechas_SubCampeon=pd.Series(['1976, 1992, 2008','1984','2000, 2012','2016','1964, 1972, 1988'], index =[3, 1, 2, 1,3],name='Año_2')
```

## Creación de dataframes 

* Un Data Frame no es otra cosa que un conjunto de pandas series que comparten un mismo índice.
* Podemos construir un Data Frame con ''concat''

* Si las series no poseen el mismo índice el resultado será un error

* Podemos intentar concatenar las series que habíamos creado, pero obtendermos un error porque los índices son diferentes

```
df = pd.concat([Ganadores_Top,N_Campeonatos,Fechas_Campeon,N_SubCampeonatos,Fechas_SubCampeon], axis=1)
```

* Para poder concatenar las series, modificamos los índices.

```
n_index =[100, 200, 300,400,500] 
# set the index
Ganadores_Top.index = n_index
N_Campeonatos.index = n_index
Fechas_Campeon.index = n_index
N_SubCampeonatos.index = n_index
Fechas_SubCampeon.index = n_index
```

* Ahora si. Concatenamos

```
df = pd.concat([Ganadores_Top,N_Campeonatos,Fechas_Campeon,N_SubCampeonatos,Fechas_SubCampeon], axis=1)
  
df
```

## Uniendo dataframes esto es para otra clase.

### NUevo data frame con columnas comunes a las de mic2

```
A=['Juan', 'Ana', 'Mili','Peter','Mario','Pepe','Rita','Sophie']
B=['Art']

mic3=pd.DataFrame(
[[5],
[4],
[3],
[4],
[3],
[2],
[1],
[5]],
index=A,
columns=B)

mic3.head()
```

```
# Dando nombre a la columna de *INDICE*
mic2.index.name = 'Nombre'
mic3.index.name = 'Nombre'
```

### *Merge*

```
# Join matching rows from mic2 to mic3.
# Incluye solo las filas que estén en mic2

pd.merge(mic2, mic3,
how='left', on='Nombre')
```

```
# Join matching rows from mic2 to mic3.
# Incluye solo las filas que estén en mic3

pd.merge(mic2, mic3,
how='right', on='Nombre')
```

```
# Join data. Retain only rows in both sets.

pd.merge(mic2, mic3,
how='inner', on='Nombre')
```

```
# Este otro Merge es una UNIÓN
pd.merge(mic2, mic3,
how='outer', on='Nombre')
```
